## 탐색
- 스택 fifo
- 큐 fifo
- 재귀함수
    - 재귀는 종료 조건 명시 필요
    - 스택 대신 재귀 사용 -> 함수 호출 연속으로 스택에 쌓임
    
- dfs
    - 깊이 우선 탐색
    - 스택 사용
        - 탐색 시작 노드를 스택에 삽입 & 방문 처리
        - 스택의 최상단 노드에 방문하지 않은 인접노드 있으면 스택에 넣고 방문 처리
        - 방문하지 않은 인접 노드 없다면 스택에서 최상단 노드 꺼내기
        - 두번째 과정 반복
        
- bfs
    - 너비 우선 탐색
    - 큐 이용
        - 탐색 시작 노드를 큐에 삽입하고 방문 처리
        - 큐에서 노드 꺼낸 뒤 해당 노드의 인접노드 방문 하지 않았다면 노드 모두 큐에 삽입
          그리고 방문 처리
        - 두번쨰 반복
        

```python
# 종료 조건 있는 재귀함수
def recursive_function(i):
    # 100번째 호출을 했을 때 종료되도록 종료 조건
    if i == 100:
        return 
    print(i, '번째 재귀함수에서', i+1, '호출')
    recursive_function(i+1)
    print(i, '번째 재귀함수 종료')


recursive_function(1)
# 1 .. 99번째에서 100번 호출 -> 99...1 종료
```

```python
# 팩토리얼 
# 반복적으로 구현한 n!
def factorial_iterative(n):
    result = 1
    # 1~n까지의 수를 차례대로 곱하기
    for i in range(i, n+1):
        result *= i
        return result

# 재귀적으로 구현한 n!
def factorial_iterative(n):
    if n<=1:
        return 1
    return n*factorial_iterative(n-1)
```

```python
# 인접리스트
# 행이 3개인 2차원 리스트로 인접리스트 표현
graph = [[] for _ in range(3)]

# 노드 0에 연결된 노드 정보 저장 (노드, 거리)
graph[0].append((1 , 7 ))
graph[0].append((2, 5))
# 노드 1에 연결된 노드 정보 저장 (노드, 거리)
graph[1].append((0, 7))

# 노드 2에 연결된 노드 정보 저장 (노드, 거리)
graph[2].append((0, 5))
print(graph)
```

```python
# dfs 함수 정의
def dfs(graph, v, visited):
    # 현재 노드를 방문 처리
    visited[v] = True
    print(v, end=' ')
    # 재귀 방문 연결노드
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i ,visited)
            
graph = [ # []은 0번 노드
  [],
  [2, 3, 8],
  [1, 7],
  [1, 4, 5],
  [3, 5],
  [3, 4],
  [7],
  [2, 6, 8],
  [1, 7]
]

visited = [False]*9

dfs(graph, 1, visited)

## 1 2 7 6 8 3 4 5  =>  스택에 들어간 순서다 , 햇갈리면 글림을 그려보자
```

```python
## bfs 함수 정의
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start]= True
    
    while queue:
        # 출력
        v = queue.popleft()
        print(v, end=' ')
        
        # 인접 노드 삽입
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i]=True

graph = [
  [],
  [2, 3, 8],
  [1, 7],
  [1, 4, 5],
  [3, 5],
  [3, 4],
  [7],
  [2, 6, 8],
  [1, 7]
]  # 1 2 3 8 7 4 5 6

# 각 노드가 방문된 정보를 리스트 자료형으로 표현(1차원 리스트)
visited = [False] * 9

bfs(graph, 1, visited)
```
