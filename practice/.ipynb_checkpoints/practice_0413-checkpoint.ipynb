```python
# 배추
import sys
sys.setrecursionlimit(10000)

def dfs(x, y):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if nx >=0 and nx<n and ny>=0 and ny<m and graph[nx][ny]==1 :
            graph[nx][ny] = 0 # 1. 방문 처리
            dfs(nx, ny) # 2. dfs 탐색
    
t = int(input())
for _ in range(t):
    # 가로(열), 세로(행), 위치개수
    m, n, k = map(int, input().split())
    graph = [[0]*m for _ in range(n)]
    result = 0
    
    for _ in range(k):
        # 위치 열, 행 순으로 받기
        b, a = map(int, input().split())
        graph[a][b] = 1
        
    for i in range(n):
        for j in range(m):
            if graph[i][j] == 1:
                dfs(i, j)
                # 한 분기에 대한 탐색이 끝났을때 카운트 하기
                result += 1
    print(result)
```

```python
# 연결요소의 개수 11724
# 인접리스트로 해결
import sys
sys.setrecursionlimit(10000)

def dfs(v):
    visited[v] = True
    for i in graph[v]:
        if not visited[i]:
            dfs(i)
    
    
n,m = map(int, input().split())
graph = [[] for _ in range(n+1)]
visited = [False]*(n+1)
cnt = 0

for _ in range(m): 
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a) # 양방향 처리


for i in range(1, n+1):
    if not visited[i]:
        dfs(i)
        cnt += 1

print(cnt)
```

```python
# 연결요소의 개수 11724
from collections import deque
import sys
sys.setrecursionlimit(10000)

def bfs(v):
    q= deque()
    q.append(v)
    visited[v] = 1
    while q:
        now = q.popleft()
        for j in graph[now]:
            if not visited[j]:
                visited[j] =1
                q.append(j)
 

n, m = map(int, input().split())
graph = [[]  for _ in range(n+1)]
visited = [0]*(n+1)

for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)

cnt = 0
for i in range(1, n+1):
    if not visited[i]:
        bfs(i)
        cnt += 1

print(cnt)
```

```python
# 로또
# 조합으로 풀자
from itertools import combinations
number_list = list(map(int, input().split()))

while True: # 엡력 테케가 정해져 있지 않고, 0을 입력하면 탈출
    try:
        number_list = sorted(number_list[1:])
        for numbers in list(combinations(number_list, 6)): # 조합 함수 사용
            for num in numbers:
                print(num, end=' ')
            print()
        print()

        number_list = list(map(int, input().split()))
        
    except EOFError:
        break
```

```python
from itertools import combinations

while True:
    number_list = list(map(int, input().split()))
    if number_list[0] == 0:
        break
    del number_list[0]
    number_list = list(combinations(number_list, 6))
#     print(number_list)
    for i in number_list:
        for j in i:
            print(j, end=' ')
        print()
    print()
```

```python
from collections import deque

def bfs(start_x, start_y, end_x, end_y):
    q = deque()
    q.append([start_x, start_y])
    
    dx = [-1, -2, -2, -1 ,1, 2, 2, 1]
    dy = [-2, -1, 1, 2, 2, 1, -1, -2]

    while q:
        x, y = q.popleft()
        
        if x == end_x and y ==end_y:
            return (graph[x][y]-1)
        
        for i in range(8):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx>=0 and nx < n and ny >= 0 and ny< n and graph[nx][ny]==0:
                graph[nx][ny] = graph[x][y] +1
                q.append([nx, ny])
    
t = int(input())

for _ in range(t):
    n = int(input())
    graph = [[0]*n for _ in range(n)]
    
    start_x, start_y = map(int, input().split())
    graph[start_x][start_y] = 1
    
    end_x , end_y = map(int, input().split())
    
    answer = bfs(start_x, start_y, end_x, end_y)
    print(answer)
```
